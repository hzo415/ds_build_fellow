```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#pandas as pd pydata documentation
#https://pandas.pydata.org/docs/user_guide/index.html
```

```python
df = pd.read_csv('HealthData.csv')
```

```python
# Plot a scatter plot between BMI and insurance charges. 
#Write down your observations about the relationship between these two variables.
```

```python
plt.figure(figsize =(10,6))
sns.scatterplot(x='bmi', y='charges', data=df, palette ='viridis', alpha = 0.7)
```

There is a slight positive linear relationship between BMI and charges
